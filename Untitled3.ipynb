1)Difference between function and method in Python?

-> Function: Independent block of code, not tied to an object.
  Method: A function defined inside a class, works on its objects.

   def add(a, b): return a+b       # function
   print("hello".upper())          # method

2)Explain function arguments and parameters in Python?

-> Parameter: Variable in function definition.
  Argument: Actual value passed.

  def greet(name): print("Hello", name)  # name = parameter
  greet("Alex")                          # "Alex" = argument

3)Ways to define and call a function in Python?

-> Functions can be written in several ways:
   Normal function: defined using def.
   With default arguments: provide default values.
   With keyword arguments: specify names in call.
   Variable arguments: *args for multiple values, **kwargs for key-value pairs.
   Anonymous functions: using lambda

            # Normal
            def f1(a, b): return a+b
            print(f1(3,4))

            # Default
            def f2(a, b=5): return a+b
            print(f2(2))  

            # Keyword
            print(f2(b=10, a=4))

            # *args
            def f3(*nums): return sum(nums)
            print(f3(1,2,3,4))

            # **kwargs
            def f4(**data): return data
            print(f4(name="Alex", age=20))

            # Lambda
            f5 = lambda x,y: x*y
            print(f5(2,3))


4)Purpose of return statement in Python?

-> Ends function execution and sends result back.
   Without return, Python returns None.

   def square(x): return x*x
   print(square(5))    # 25

5)What are iterators? How different from iterables?

-> Iterable: Any object we can loop over (list, tuple, string).
   Iterator: Produces values one by one using next().

   lst=[1,2,3]
   it=iter(lst)
   print(next(it))   # 1
   print(next(it))   # 2

6)Concept of generators and how to define them?

-> Functions that use yield instead of return.
   Generate values lazily (only when needed).
   
   def gen_nums():
    yield 1
    yield 2
    yield 3
   print(list(gen_nums()))

7)Advantages of generators over functions?

-> Memory efficient (no need to store entire list).
   Can represent infinite sequences.
   Easier to read than building custom iterators.

8)What is a lambda function? When used?

-> One-line anonymous function, usually used for small tasks

   square = lambda x: x*x
   print(square(6))

9)Purpose and usage of map() in Python?

-> Applies a function to each element of an iterable.

   nums=[1,2,3,4]
   print(list(map(lambda x:x*2, nums)))

10)Difference between map(), reduce(), and filter()?

-> map(): transforms each item.
   reduce(): accumulates into one value.
   filter(): selects items meeting condition.

    from functools import reduce
    nums=[1,2,3,4]
    print(list(map(lambda x:x*2, nums)))       # [2,4,6,8]
    print(list(filter(lambda x:x%2==0, nums))) # [2,4]
    print(reduce(lambda a,b:a+b, nums))        # 10

```python
# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return self.a


print(list(Fibonacci(10)))
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

x=power_of_two(5)
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```

```python
# Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    with open(filename) as f:
        for line in f:
            yield line.strip()

for line in read_file_lines("example.txt"):
    print(line)
```

```python
# Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 3), (2, 1), (3, 2)]
data.sort(key=lambda x: x[1])
print(data)
```

```python
# Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

celsius = [0, 20, 30, 40]
fahrenheit = list(map(lambda x: (x * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
# Create a Python program that uses filter() to remove all the vowels from a given string.

s = "Hello World"
result = list(filter(lambda x: x.lower() not in "aeiou ", s))
print(result)
```
